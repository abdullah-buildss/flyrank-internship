```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.


# Quick check - 3 numbers that support my lane
import pandas as pd
import json

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(f"✅ Data loaded! {len(df)} rows")

# Number 1: How many pages are declining?
declining = (df["trend_direction"].str.lower() == "down").mean()
print(f"\n1. Pages declining: {declining:.1%}")

# Number 2: How many pages are stale?
stale = (df["days_since_last_update"] >= 180).mean()
print(f"2. Pages not updated in 6+ months: {stale:.1%}")

# Number 3: ML vs hand rule
try:
    with open("outputs/model_results.json", "r") as f:
        res = json.load(f)
    hand = res["baseline"]["baseline_precision_at_50"]
    ml = res["models"]["random_forest"]["precision_at_50"]
    print(f"3. ML Precision@50: {ml:.3f} vs Hand rule: {hand:.3f} ({ml/hand:.1f}x better)")
except FileNotFoundError:
    print("3. Run the pipeline first: !python scripts/run_all.py")
```
