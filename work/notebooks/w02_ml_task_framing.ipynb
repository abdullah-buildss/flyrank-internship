## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*



**Simple rules don't work well because:**

1. **The data proves it:** Hand rule Precision@50 = 0.240 vs ML = 0.740 (3.1x better)
2. **Multiple signals interact** - a page might have good CTR but bad position. Rules can't capture these combos.
3. **Patterns are non-linear** - the relationship between signals and decline isn't always straight. ML finds these patterns.
4. **ML adapts** - if data changes, ML can be retrained. Fixed rules need manual updates.

**Bottom line:** If a hand-written rule could solve this perfectly, we wouldn't need ML.
